```python
import pandas as pd
full_df = pd.read_csv('../../Data/SMTO_2015/SMTO_2015_Complete_Input.csv')
full_df = full_df[full_df['Level'] != 'Other']
full_df.head()
```

```python
full_df[['PD', 'Family']].corr()
```

```python
school_codes = list(full_df['School'].unique())

def row_to_region(x):
    pd = x.PD
    if pd == 1:
        return 'Downtown'
    elif pd <= 6:
        return 'Midtown'
    elif pd <= 16:
        return 'Uptown'
    elif x.HomeZone < 4000:
        return 'Inner Suburbs' # Durham, York, Peel
    else:
        return 'Outer Suburbs' # Halton, Hamilton
```

```python
for level in (['UG', ' '], ['Grad'], ['UG', 'and', 'Grad']):
    df = full_df[full_df['Level'].isin(level)]
    counts = df['PD'].value_counts()
    PDs = counts[counts >= 10].index
    print("---------------", *level, "(" + str(len(PDs)) + " districts) ---------")
    df = df[df['PD'].isin(PDs)]
    temp = df[['Dist.' + code for code in school_codes]].rank(axis=1)
    series = pd.DataFrame('Dist.' + df['School'])
    series['Index'] = series.index
    df['Dist_Rank'] = series.apply(lambda x: temp[x.School][x.Index], axis=1)
    #print(df.groupby('PD')['Dist_Rank'].mean())
    
    df['Region'] = df.apply(row_to_region, axis=1)
    a = df.groupby('Region')['Dist.SG'].mean()
    b = df.groupby('Region')['Family'].mean()
    c = df.groupby('Region')['Dist_Rank'].mean()
    df_to_print = pd.concat((a, b, c), axis=1).sort_values('Dist.SG')
    print(df_to_print)
    print()
    
    corrs = df_to_print.corr().values
    print("Corr DistSG/Dist_Rank: {:.3f}".format(corrs[0][1]))
    print("Corr Family/Dist_Rank: {:.3f}".format(corrs[1][2]))
    print("\n")
```
