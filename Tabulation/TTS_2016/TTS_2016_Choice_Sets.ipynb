```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../../Data/TTS_2016/TTS_2016_Subset.csv', index_col=0)
```

```python
fixed_codes = pd.read_csv('../../Data/TTS_2016/SubsetSchoolCodes.csv')
fixed_codes['SchoolName'] = fixed_codes['Name'].apply(lambda x: x.split(' --')[0].strip())
fixed_codes = fixed_codes.set_index('SchoolName')
df['Campus'] = df['SchoolName'].apply(lambda x: fixed_codes.loc[x]['Short'])
df['Inst'] = df['SchoolName'].apply(lambda x: fixed_codes.loc[x]['Inst'])
df['Grad/Adult'] = df['SchoolName'].apply(lambda x: bool(fixed_codes.loc[x]['Grad/Adult']))
df['Inst'].value_counts()
```

```python
unis = ('UT', 'YK', 'RY', 'MC', 'IT', 'OC', 'GU', 'BR', 'TR','TY')
private = ('TY', 'TS')
df['Private'] = df['Inst'].isin(private)
df['University'] = df['Inst'].isin(unis)
```

```python
print(df['Distance'].describe())

numBins = 500
plt.figure()
df['Distance'].plot.hist(bins=numBins)
plt.xlabel('Distance (km)')
plt.show()

boxplot = df.boxplot(column=['Distance'])
```

The resutls show that on average students commute around __16 km__ every day. The spread of the data is large, however, which is normal because here we are not differentiating between the different modes of transportation that students may take. Those who walk, for example, will travel a lot less than 16 km to get to school while those who take regional transit will probably travel a lot more than that. We can see that there are also many outliers in the data who suposedly travel very long distances every day. The maximum distance travelled from this data set is around 120 km, which is more or less the distance between Whitby and Hamilton. It is highly unlikely for a person to be travelling this distance to get to school, which makes us question the reliability of this data. 

#### Distance per Mode of Transportation

To better analyzer our data, we divided the distance distributions among the diferent modes of transportation present in the survey:

```python
fig = plt.figure(figsize=(16, 8))
labels=['Public Transit', 'Auto Driver', 'Auto Passenger', 'Walk', 'GO Rail + Public Transit', 'Bicycle', 'GO Rail', 'Paid Rideshare', 'Taxi', 'Motorcycle', 'Other']
modes = ('B','D','P','W','J','C','G','U','T','M','O')
for i in range(len(modes)):
    temp_df = df.loc[df['ModeTakenToSchool'] == modes[i]]
    plt.subplot(2, 6, i+1)
    boxplot = temp_df.boxplot(column=['Distance'])
    plt.title(labels[i])
    plt.ylim((0, 125000))
    plt.tight_layout()
plt.show()
```

In general, the average distance of every mode of transportation makes sense. Those taking public transit travel around __10 km__, those who walk travel close to __1 km__, while those who take the GO Rail travel more tahn __30 km__. The main concern when looking at this data is the amount of respondents who claim to travel relatively long distances in certain modes of transportation. There are some people who entered a walking distance of about 90 km, when walking more than 15 km is already unreasonable. A similar thing happens with cyclists since some claim to to travel up to 60 km. Even for public transit there are too many outliers who say they commute more than 60 km, which is not likely. It is also unlikely for people to drive to school more than 100 km every day. It is possible that we may need to filter this data to remove a lot of these obvious outliers. 

## 2. Auto Travel Time

Next, we examine the auto travel time between the home and school zones of the respondents:

```python
print(df['AutoTravelTime'].describe())
numBins = 500
plt.figure()
df['AutoTravelTime'].plot.hist(bins=numBins)
plt.xlabel('Auto Travel Times')
plt.show()
```

The resutls are not too outrageous with an average of __27 minutes__. However, this auto travel time between traffic zones includes the respondents who use every type of mode, which explains the large spread inthe data.

#### Looking only at Car Users
It may be better to look only at the auto travel times for those respondents that travel to school by car:

```python
drive_df = df.loc[df['ModeTakenToSchool'].isin(['D','P','T','U'])]
print(drive_df['AutoTravelTime'].describe())

boxplot = drive_df.boxplot(column=['AutoTravelTime'])
```

Interestingly, the spread of the data does not change much from above. In fact, we still have many outliers who apparently travel more than 2 hours by car. 

## 3. Transit Travel Time
Next, we examine the transit travel times between the home and school zones of the respondents. This time is estimated by adding the average walking time from a student's home zone to the nearest transit station, the average time spent waiting for the transit, the average time spent in the transit vehicle, and the average time it takes to walk from the arrival station to the school zone.

```python
print(df['TransitTravelTime'].describe())
numBins = 500
plt.figure()
df['TransitTravelTime'].plot.hist(bins=numBins)
plt.xlabel('Transit Travel Times')
plt.show()
```

As we can see, these transit travel times seem very high in general. The average transit travel time is of __1 hour 25 minutes__, which and there is a large number of transit travel times above 2 hours and up to 7.5 hours. Similar to what we did for auto travel times, it might be beneficial to look at transit travel time for people who only take transit.

#### Looking only at Public Transit Users

Let's look only at the population of respondents who only take public transit (not including GO Rail):

```python
transit_df = df.loc[df['ModeTakenToSchool'] == 'B']
transit_df['TransitTravelTime'].describe()
```

```python
boxplot = transit_df.boxplot(column=['TransitTravelTime'])
```

The average and spread still look to high, so it is worth dividing this transit travel time into a) walking, b) waiting and c) in-vehicles components, to see which part contributes most to this abnormal transit time:

__a) Walking time:__

```python
boxplot = transit_df.boxplot(column=['WalkingTime'])
```

We can see that the average walking time of __20 minutes__ is not bad, but there are some very evident outliers. Walking to and from transit for more than 40 minutes seems a little unlikely. It is possible that these numbers are so high because the program used to calculate the walking times was not cosidering that some students are driven to their transit station, for example.

__b) Waiting Time:__

```python
boxplot = transit_df.boxplot(column=['WaitingTime'])
```

Waiting time looks reasonable. 

__c) In-Vehicle Time:__

```python
boxplot = transit_df.boxplot(column=['tivtt'])
```

In-vehicle time look a bit suspicious. Even though the average of __50 minutes__ is possible, there is a high number of students who spend more than an hour in their transit vehicle, which does not seem very reasonable considering that this only includes buses, subways, and streetcars. 

#### Looking at GO Rail Users Only

Let's now look at the transit travel times of those students who take the GO Rail to school.

```python
df_GO = df.loc[df['ModeTakenToSchool'].isin(['G','J'])]
boxplot = df_GO.boxplot(column=['TransitTravelTime'])
```

Once again, the times we get are very high. We now have an average of around __2 hours__ of transit commute with some very high outlier, which is definitely not accurate. Let's divide the time again into  a) walking, b) waiting and c) in-vehicles components, to see which part contributes most to this abnormal transit time:

__a) Walking Time:__

```python
boxplot = df_GO.boxplot(column=['WalkingTime'])
```

Walking time is possible but quite high once again considering the large number of people walking for more than an hour. Again, it is possible that these times are so high because the program used to calculate them did not consider that some students are driven to their transit station.


__b) Waiting Time:__

```python
boxplot = df_GO.boxplot(column=['WaitingTime'])
```

Waiting time, again, seems reasonable. 

__c) In-Vehicle Time:__

```python
boxplot = df_GO.boxplot(column=['tivtt'])
```

These in-vehicle times also seem very off. It looks like students taking the GO Rail spend more than an hour inside of the train on average. From personal experience, this does not sound accurate at all.
