# Mode Choice Tabulation - StudentMoveTO 2015

In this notebook, I tabulate the mode choices of respondents of the 2015 StudentMoveTO survey.

Mode choices are taken from the _psmainmodefalltypicalaggr_ column in the Respondents data file. The column has been added to the _Segmented.xlsx_ file. This column contains the mode choice each respondent identified as their main mode used to commute to campus on a typical fall day, based on distance. The possible values contained in this column are:

- Walk: Walking  
- Bike: Cycling  
- Local Transit: Bus, Streetcar, Subway/RT  
- Regional Transit: GO Bus, GO Train
- Solo Driver: Driver alone  
- Rideshare: Driver with passenger, Car passenger  
- Other: Motorcycle, Moped, Intercampus Shuttle, Paratransit, Taxi, Other
- Do Not Travel: Distance learners

Below, I explore the mode choice distribution for the entire sample as well as for individual segments. These segments are based off of the segmentation process completed previously. Then, I explore the relationship between mode choice and distance.

First, let us load the data into a Pandas DataFrame.

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('../Data/SMTO_2015/SMTO_2015_Segmented.csv')
df = df.dropna()  # Remove 227 rows with empty values
df.head()
```

## Mode Choice Distribution

First, let us see the mode choice distribution in the entire dataset.

```python
print(df['psmainmodefalltypicalaggr'].value_counts())
print("\n")
print(df['psmainmodefalltypicalaggr'].value_counts(normalize=True))
print("\n")

plt.figure(figsize=(16,6))
plt.subplot(1,2,1)
plt.title('Mode choice frequency for entire dataset')
df['psmainmodefalltypicalaggr'].value_counts().sort_index().plot(kind='bar')
plt.subplot(1,2,2)
df['psmainmodefalltypicalaggr'].value_counts().sort_index().plot.pie(labeldistance=None)
plt.legend(bbox_to_anchor=(-0.2, -0.2), loc="lower left")
plt.title('Mode choice distribution for entire dataset')
plt.show()
```

Local transit is clearly the most popular mode choice, with about half of respondents identifying this as their typical mode. 19% of respondents walk to campus and 13% use regional transit. Smaller proportions of respondents bike or drive to campus.

Now, let us take a look at the distribution of mode choices by segment. Recall that the segments are labelled as follows:

| Level | Full Time | | Part Time | |
| --- | --- | --- | --- | --- |
| | Family | Not | Family | Not |
| First-year UG | 1 | 2 | 3 | 4 |
| Upper-year UG | 5 | 6 | 7 | 8 |
| Grad | 9 | 10 | 11 | 12 |
| Other | 0 | 0 | 0 | 0 |

```python
fig = plt.figure(figsize=(16, 16))
for i in range(13):
    temp_df = df.loc[df['Segment'] == i]    
    plt.subplot(3, 5, i+1)
    plt.tight_layout()
    plt.title('Segment ' + str(i))
    temp_df['psmainmodefalltypicalaggr'].value_counts(normalize=True).sort_index().plot(kind='bar')
plt.show()

fig = plt.figure(figsize=(16,12))
for i in range(13):
    temp_df = df.loc[df['Segment'] == i]    
    plt.subplot(3, 5, i+1)
    plt.title('Segment ' + str(i))
    temp_df['psmainmodefalltypicalaggr'].value_counts().sort_index().plot.pie(labeldistance=None)

plt.legend(bbox_to_anchor=(2, 0.15), loc="lower right")
plt.show()
```

Note that in the pie charts above the colours in the legend are not true for all charts. Segments with no counts of a certain mode will not have that mode represented so it will have been skipped. I am working on fixing this issue.

Clearly, the mode choice distribution fluctuates significantly between segments. For instance, full-time undergraduate students list walking as their most popular mode choice, whereas other segments list local transit as their most popular mode choice.

One possible next step would be to perform equivalence tests to establish whether our segmentation approach was justified - that is, whether different segments had statistically significant differences in mode choice distribution.

## Relationship Between Mode Choice and Distance

In this section, let's take a look at how mode choice is correlated with distance from school to campus. This distance is stored in the column _HomeToMainCampusKM_.

First, let's make a boxplot of the data in this column.

```python
boxplot = df.boxplot(column=['HomeToMainCampusKM'])
```

Interestingly, we see that there are many students whose distances are listed upwards of 1000km from the campus they attend. Such values could be the result of distance learners or out-of-town students who have listed their permanent address instead of where they live while attending campus. Let us filter out rows with distance values of over 130km, since these likely do not accurately reflect a respondents' actual commute.

```python
df.drop(df[df['HomeToMainCampusKM'] > 130].index, inplace=True)
boxplot = df.boxplot(column=['HomeToMainCampusKM'])
df['HomeToMainCampusKM'].describe()
```

This looks like a much more reasonable spread. Notably, the mean commute distance is about 13.5km, while a quarter of respondents live within 2.5km of their campus.

Now, let us look at the distance distributions for various modes.

```python
fig = plt.figure(figsize=(16, 8))
modes = ('Walk', 'Bike', 'Local Transit', 'Regional Transit', 'Solo Driver', 'Rideshare', 'Other')
for i in range(len(modes)):
    temp_df = df.loc[df['psmainmodefalltypicalaggr'] == modes[i]]
    plt.subplot(1, 7, i+1)
    boxplot = temp_df.boxplot(column=['HomeToMainCampusKM'])
    plt.title(modes[i])
    plt.ylim((0, 125))
    plt.tight_layout()
plt.show()
```

From the boxplots above, we see several predictable trends. Firstly, the median commute distance increases in the following order: Walk > Bike > Local Transit > Regional Transit, and Rideshare > Solo Driver.

For the most part, distances for walk and bike trips seem reasonable. However, there are a number of respondents whose listed mode was walking, but who commute upwards of 20km. This suggests that the listed mode for some respondents does not accurately reflect the majority of their commute, or that respondents listed different locations for their home than where they actually commute to/from. The degree to which this problem occurs is interestingly much smaller for bikers.
