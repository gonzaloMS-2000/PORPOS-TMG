```python
import pandas as pd
import math
import csv
```

```python
df = pd.read_csv('../../Data/SMTO_2015/SMTO_2015_Complete_Input.csv')
df.head()
```

```python
sample_props = df['School_Codes'].value_counts(normalize = True)
```

```python
sample_props
```

```python
preds_HZ = pd.read_csv('../Map Projects/AverageStudentPopulation - Needed Files/RF_Probabilities_HZ.csv')
del preds_HZ['Origin']
preds_HZ
```

```python
preds_dist = pd.read_csv('../Map Projects/AverageStudentPopulation - Needed Files/RF_Probabilities_dist.csv')
del preds_dist['Origin']
preds_dist
```

```python
RF_Probs = pd.DataFrame()
```

```python
for col in preds_HZ.columns:
    RF_Probs[col + "_Norm_HZ"] = preds_HZ[col] / sample_props[col]
for col in preds_dist.columns:
    RF_Probs[col + "_Norm_dist"] = preds_dist[col] / sample_props[col]
```

```python
RF_Probs
```

```python
RF_Probs.to_csv('../Map Projects/AverageStudentPopulation - Needed Files/RF_Norm_Schools.csv', index = False)
```
