```python
df = pd.read_csv('../../Data/SMTO_2015/SMTO_2015_Complete_Input.csv')
df.head()
```

```python
map_df = pd.DataFrame()
```

```python
zones = pd.read_csv('../Needed_Zones.csv')
map_df['Origin'] = zones['Origin']
map_df.set_index(map_df['Origin'],inplace = True)
```

```python
map_df
```

```python
# Sample - Real Schools
def top_school(x):
    try:
        return df[df['HomeZone'] == x]['School_Codes'].value_counts().index[0]
    except:
        return None
    
map_df['Sample_TopSchool'] = map_df['Origin'].apply(top_school)
```

```python
def school_prob(x,code):
    temp = df[df['HomeZone'] == x]['School_Codes'].value_counts(normalize = True)
    try:
        return temp[code]
    except:
        return 0 if len(temp) else None

codes = list(df['School_Codes'].unique())
for code in codes:
    map_df[code + '_Sample_Prob'] = map_df['Origin'].apply(lambda x: school_prob(x, code)) / df['School_Codes'].value_counts(normalize=True)[code]

map_df
```

```python
sample_props = df['School_Codes'].value_counts(normalize = True)
sample_props
```

```python
# Gravity - mlogit
preds_grav = pd.read_csv('../Map Projects/AverageStudentPopulation - Needed Files/mlogit_prediction.csv')
preds_grav
#preds_grav.set_index(zones['Origin'])
```

```python
map_df['Grav_Pred'] = preds_grav.idxmax(axis = 1)
for col in preds_grav.columns:
    map_df[col + "_Norm_grav"] = preds_grav[col] / sample_props[col]
```

```python
map_df
```

```python
# RFs - Predictions
rf_df = pd.read_csv('../../Data/SMTO_2015/RF_Predictions.csv')
rf_df.tail()
```

```python
# RF - HomeZone - Norms
preds_HZ = pd.read_csv('../Map Projects/AverageStudentPopulation - Needed Files/RF_Probabilities_HZ.csv')
del preds_HZ['Origin']
preds_HZ
```

```python
map_df['HZ_Pred'] = rf_df['HZ']
for col in preds_HZ.columns:
    map_df[col + "_Norm_HZ"] = preds_HZ[col] / sample_props[col]
```

```python
# RF - Dist - Norms
preds_dist = pd.read_csv('../Map Projects/AverageStudentPopulation - Needed Files/RF_Probabilities_dist.csv')
del preds_dist['Origin']
preds_dist
```

```python
map_df['Dist_Pred'] = rf_df['Dists']
for col in preds_dist.columns:
    map_df[col + "_Norm_dist"] = preds_dist[col] / sample_props[col]
```
