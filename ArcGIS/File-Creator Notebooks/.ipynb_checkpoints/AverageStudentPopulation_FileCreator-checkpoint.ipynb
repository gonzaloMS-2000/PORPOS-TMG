```python
import pandas as pd
import math
import csv
import numpy as np
```

```python
df = pd.read_csv('../../Data/SMTO_2015/SMTO_2015_Complete_Input.csv')
df.head()
```

```python
map_df = pd.DataFrame()
```

```python
zones = pd.read_csv('../Needed_Zones.csv')
map_df['Origin'] = zones['Origin']
map_df.set_index(map_df['Origin'],inplace = True)
```

```python
map_df
```

```python
# Sample - Real Schools

#sample_df = pd.DataFrame()
#sample_df['Origin'] = df['HomeZone']
#sample_df['School'] = df['School_Codes']

homezones = set(df['HomeZone'].value_counts().index.tolist())
zone_to_campus = {}

for i in homezones:
    SG = ((df['School_Codes'] == 'SG') & (df['HomeZone'] == i)).sum()
    YK = ((df['School_Codes'] == 'YK') & (df['HomeZone'] == i)).sum()
    RY = ((df['School_Codes'] == 'RY') & (df['HomeZone'] == i)).sum()
    SC = ((df['School_Codes'] == 'SC') & (df['HomeZone'] == i)).sum()
    MI = ((df['School_Codes'] == 'MI') & (df['HomeZone'] == i)).sum()
    OC = ((df['School_Codes'] == 'OC') & (df['HomeZone'] == i)).sum()
    YG = ((df['School_Codes'] == 'YG') & (df['HomeZone'] == i)).sum()

    x_names = ['SG','YK','RY','SC','MI','OC','YG']
    x = [SG,YK,RY,SC,MI,OC,YG]
    x_num = [1,2,3,4,5,6,7] 
    
    zone_to_campus[i] = x_names[x.index(max(x))]
    
top_school_df = df.from_dict(zone_to_campus,orient='index')
print(top_school_df.shape)
top_school_df = top_school_df.rename(columns = {0:'code'})
top_school_df.head()
```

```python
not_found = set()
map_df['Sample_Top_School'] = np.nan

for i in top_school_df.index.tolist():
    try:
        map_df['Sample_Top_School'][i] = top_school_df['code'][i]
    except:
        not_found.add(i)
```

```python
sample_props = df['School_Codes'].value_counts(normalize = True)
sample_props
```

```python
# Gravity - mlogit
preds_grav = pd.read_csv('../Map Projects/AverageStudentPopulation - Needed Files/mlogit_prediction.csv')
preds_grav
```

```python
map_df['Grav_Pred'] = preds_grav.idxmax(axis = 1)
for col in preds_grav.columns:
    map_df[col + "_Norm_grav"] = preds_grav[col] / sample_props[col]
```

```python
# RFs - Predictions
rf_df = pd.read_csv('../../Data/SMTO_2015/RF_Predictions.csv')
rf_df.tail()
```

```python
# RF - HomeZone - Norms
preds_HZ = pd.read_csv('../Map Projects/AverageStudentPopulation - Needed Files/RF_Probabilities_HZ.csv')
del preds_HZ['Origin']
preds_HZ
```

```python
map_df['HZ_Pred'] = rf_df['HZ']
for col in preds_HZ.columns:
    map_df[col + "_Norm_HZ"] = preds_HZ[col] / sample_props[col]
```

```python
# RF - Dist - Norms
preds_dist = pd.read_csv('../Map Projects/AverageStudentPopulation - Needed Files/RF_Probabilities_dist.csv')
del preds_dist['Origin']
preds_dist
```

```python
map_df['Dist_Pred'] = rf_df['Dists']
for col in preds_dist.columns:
    map_df[col + "_Norm_dist"] = preds_dist[col] / sample_props[col]
```

```python
map_df.to_csv('../Map_Inputs.csv', index = False)
```
