## Threshold for Closest School Dummies

Previously, we've seen some improvements over both the "Proposed" and "Eric" models when introducing a dummy variable for the closest school to each students' home zone. In this notebook, we explore whether including this dummy only for distances under a certain threshold can further improve our results.

First, let's load the data.

```python
# Import necessary packages
import pandas as pd
import csv

# Load data and remove Other students
df = pd.read_csv('../../../Data/SMTO_2015/SMTO_2015_Complete_Input.csv')
df = df[df['Level'] != 'Other']

# Convert school column to numeric
school_codes = df.School.unique().tolist()
df['School'] = df['School'].apply(lambda x: school_codes.index(x))

# Enrollment columns based on student's level
for code in school_codes:
    df['Enrol.' + code] = df.apply(lambda x: x['UG.' + code] if x['Level'] == 'UG' else x['Grad.' + code], axis=1)
    
num_rows = df.shape[0]
```

Now, we define several useful functions for calculating metrics.

```python
# Function for confusion matrices
def get_cm(probs, hardmax):
    cm = []
    if hardmax:
        for school in range(len(school_codes)):
            cm.append([(probs[df['School'] == school][['Prob.' + i for i in school_codes]].idxmax(axis = 1) == 'Prob.' + j).sum() for j in school_codes])
    else:
        for school in range(len(school_codes)):
            cm.append((probs[df['School'] == school][['Prob.' + i for i in school_codes]].sum().values.tolist()))    
    return cm

# Function for accuracy from confusion matrix
def get_accuracy(cm):
    correct = sum([cm[i][i] for i in range(len(cm))])
    return correct/sum(sum(cm,[])) * 100
```

Now, we iterate over possible threshold values and store the metrics for each model. We begin with the Eric model since this will allow us to examine the results for statistical significance of the closest dummy for different schools.

```python
# Prepare lists for graphing
import numpy as np
ts = list(np.arange(30, 10, -2)) + list(np.arange(10, 5, -1)) + list(np.arange(5, -0.5, -0.5))
num_closests, ACCs, APOs = [], [], []

# Import Biogeme modules
import biogeme.database as db
import biogeme.biogeme as bio
import biogeme.models as models
from biogeme.expressions import Beta

# Iterate over thresholds
for t in ts:
    # Prepare dummy columns
    for code in school_codes:
        df['Closest.' + code] = ((df['Closest.' + code]) & (df['Dist.' + code] <= t)) * 1
    num_closests.append(df[['Closest.' + code for code in school_codes]].sum().sum()) 
    database = db.Database("SMTO", df.select_dtypes(include = 'number'))
    
    # Specify model
    B_CLOSESTs, V, av = [], {}, {}
    B_DIST = Beta('B_DIST', 0, None, None, 0)
    B_FAM_DIST = Beta('B_FAM_DIST', 0, None, None, 0)
    for i in range(len(school_codes)):
        code = school_codes[i]
        B_CLOSESTs.append(Beta('B_CLOSEST_' + code, 0, None, None, 0))
        V[i] = B_CLOSESTs[i] * database.variables["Closest." + code] + database.variables["Enrol." + code] + database.variables['Dist.' + code] * (B_DIST + B_FAM_DIST * database.variables["Family"])
        av[i] = 1   
    
    # Run model
    logprob = models.loglogit(V, av, database.variables["School"])
    biogeme = bio.BIOGEME(database, {'loglike': logprob, 'weight': database.variables["Exp_Segment"]})
    biogeme.modelName = "Eric_Closest_7_T=" + str(t)
    betas = biogeme.estimate().getBetaValues()    
    
    # Generate probabilities
    simulate = {'Prob.' + school_codes[i]: models.logit(V, av, i) for i in range(len(school_codes))}
    sim_biogeme = bio.BIOGEME(database, simulate)
    probs = sim_biogeme.simulate(betas).set_index(df.index)

    # Confusion matrices and metrics
    hard_cm = get_cm(probs, True)
    soft_cm = get_cm(probs, False)
    ACCs.append(get_accuracy(hard_cm))
    APOs.append(get_accuracy(soft_cm))
```

First, let's plot the number of students for which the closest dummy is included for each threshold.

```python
import matplotlib.pyplot as plt
plt.scatter(ts, num_closests)
plt.plot([0,  30], [num_rows, num_rows], '--')
plt.title("# True Dummies")
plt.xlabel('Threshold (km)')
plt.ylabel('Students')
plt.legend(labels = ['No Threshold'])
plt.show()
```

Now, let's plot the hardmax accuracies and softmax APOs for different thresholds.

```python
plt.subplots(1, 2, figsize=(12, 4))

# Plot hardmax accuracies
plt.subplot(1, 2, 1)
plt.scatter(ts, ACCs)
plt.plot([0, 30], [49.44, 49.44], '--')
plt.plot([0, 30], [49.30, 49.30], '--')
plt.title("Hardmax Accuracy")
plt.xlabel('Threshold (km)')
plt.ylabel('Accuracy')
plt.legend(labels = ['No Dummies', 'No Threshold'])

# Plot softmax APOs
plt.subplot(1, 2, 2)
plt.scatter(ts, APOs)
plt.plot([0, 30], [36.98, 36.98], '--')
plt.plot([0, 30], [37.41, 37.41], '--')
plt.title("Softmax APOs")
plt.xlabel('Threshold (km)')
plt.ylabel('APO')
plt.legend(labels = ['No Dummies', 'No Threshold'])

plt.tight_layout()
plt.show()
```

It seems that the APO is highest for low thresholds. Let's re-plot that section of the graphs:

```python
plt.subplots(1, 2, figsize=(12, 4))

# Plot hardmax accuracies
plt.subplot(1, 2, 1)
plt.scatter(ts[-10:], ACCs[-10:])
plt.plot([0, 5], [49.44, 49.44], '--')
plt.plot([0, 5], [49.30, 49.30], '--')
plt.title("Hardmax Accuracy")
plt.xlabel('Threshold (km)')
plt.ylabel('Accuracy')
plt.legend(labels = ['No Dummies', 'No Threshold'])

# Plot softmax APOs
plt.subplot(1, 2, 2)
plt.scatter(ts[-10:], APOs[-10:])
plt.plot([0, 5], [36.98, 36.98], '--')
plt.plot([0, 5], [37.41, 37.41], '--')
plt.title("Softmax APOs")
plt.xlabel('Threshold (km)')
plt.ylabel('APO')
plt.legend(labels = ['No Dummies', 'No Threshold'])

plt.tight_layout()
plt.show()
```

It seems that the best performing model is the one with a threshold of 2km. This model offers about a 0.8% improvement in accuracy and over 2.0% improvement in APO relative to the model with no dummies. Let's look at the estimated parameters.

```python
from IPython.display import HTML

best_threshold = 2.0
HTML(filename= "Eric_Closest_7_T=" + str(best_threshold) + ".html")
```

As we can see, all of the `B_CLOSEST` coefficients are positive and statistically significant. This is encouraging!

Now, let's check if we obtain similar results for the proposed model, where only one `B_CLOSEST` is estimated for all seven campuses. We need to first reload the data.

```python
df = pd.read_csv('../../../Data/SMTO_2015/SMTO_2015_Complete_Input.csv')
df = df[df['Level'] != 'Other']
school_codes = df.School.unique().tolist()
df['School'] = df['School'].apply(lambda x: school_codes.index(x))
for code in school_codes:
    df['Enrol.' + code] = df.apply(lambda x: x['UG.' + code] if x['Level'] == 'UG' else x['Grad.' + code], axis=1)
num_rows = df.shape[0]
```

Now let's run the model.

```python
ACCs, APOs = [], []

for t in ts:
    # Prepare dummy columns
    for code in school_codes:
        df['Closest.' + code] = ((df['Closest.' + code]) & (df['Dist.' + code] <= t)) * 1
    database = db.Database("SMTO", df.select_dtypes(include = 'number'))
    
    # Specify model
    V, av = {}, {}
    B_DIST = Beta('B_DIST', 0, None, None, 0)
    B_CLOSEST = Beta('B_CLOSEST', 0, None, None, 0)
    B_FAM_DIST = Beta('B_FAM_DIST', 0, None, None, 0)
    for i in range(len(school_codes)):
        code = school_codes[i]
        V[i] = B_CLOSEST * database.variables["Closest." + code] + database.variables["Enrol." + code] + database.variables['Dist.' + code] * (B_DIST + B_FAM_DIST * database.variables["Family"])
        av[i] = 1    
    
    # Run model
    logprob = models.loglogit(V, av, database.variables["School"])
    biogeme = bio.BIOGEME(database, {'loglike': logprob, 'weight': database.variables["Exp_Segment"]})
    biogeme.modelName = "Prop_Closest_7_T=" + str(t)
    betas = biogeme.estimate().getBetaValues()    
    
    # Generate probabilities
    simulate = {'Prob.' + school_codes[i]: models.logit(V, av, i) for i in range(len(school_codes))}
    sim_biogeme = bio.BIOGEME(database, simulate)
    probs = sim_biogeme.simulate(betas).set_index(df.index)

    # Confusion matrices and metrics
    hard_cm = get_cm(probs, True)
    soft_cm = get_cm(probs, False)
    ACCs.append(get_accuracy(hard_cm))
    APOs.append(get_accuracy(soft_cm))
```

Now, let's plot the hardmax accuracies and softmax APOs for different thresholds.

```python
plt.subplots(1, 2, figsize=(12, 4))

# Plot hardmax accuracies
plt.subplot(1, 2, 1)
plt.scatter(ts, ACCs)
plt.plot([0, 30], [49.59, 49.59], '--')
plt.plot([0, 30], [50.35, 50.35], '--')
plt.title("Hardmax Accuracy")
plt.xlabel('Threshold (km)')
plt.ylabel('Accuracy')
plt.legend(labels = ['No Dummies', 'No Threshold'])

# Plot softmax APOs
plt.subplot(1, 2, 2)
plt.scatter(ts, APOs)
plt.plot([0, 30], [36.79, 36.79], '--')
plt.plot([0, 30], [37.09, 37.09], '--')
plt.title("Softmax APOs")
plt.xlabel('Threshold (km)')
plt.ylabel('APO')
plt.legend(labels = ['No Dummies', 'No Threshold'])

plt.tight_layout()
plt.show()
```

This time, it seems that the APO is highest for low thresholds, whereas the accuracy is highest for high thresholds. Since we are prioritizing APO over accuracy, let us look at the results for the model with a threshold of 2km, as for above. The model with this threshold offers about 0.4% decrease in accuracy but almost 1.6% improvement in APO compared to the model with no dummies. Let's look at the estimated parameters.

```python
HTML(filename= "Prop_Closest_7_T=" + str(best_threshold) + ".html")
```

Once again, the parameters have the expected signs and are statistically significant.
