```python
import pandas as pd
df = pd.read_csv('TTS_2016_Filtered.csv', index_col = 0)
df
```

```python
# Convert codes to names
codes_df = pd.read_csv('FullSchoolCodes.csv', index_col=0)
df['SchoolName'] = df['SchoolCode'].apply(lambda x: codes_df.loc[x]['Name']).apply(lambda x: x[:-4] if x.endswith('--') else x)
```

```python
# Manually matched some schools using SchoolCodes - Copy.csv
fixed_codes = pd.read_csv('SubsetSchoolCodes.csv', index_col=0)
fixed_codes.head()
print(df[~df['SchoolCode'].isin(fixed_codes.index)]['SchoolName'].value_counts().head(20))
df[~df['SchoolCode'].isin(fixed_codes.index)]['SchoolCode'].value_counts().head(20)
# Largest remaining schools
```
